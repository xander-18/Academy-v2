## Definimos una variable para calcular la media


## average_media = []

```python
import numpy as np

matriz = np.random.randint(0, 10, size=(5, 5))
print("Matriz:")
print(matriz)


media = np.mean(matriz)
desviacion_estandar = np.std(matriz)

print("Media:", media)
print("Desviación estándar:", desviacion_estandar)

```

```python
import numpy as np
import matplotlib.pyplot as plt


matriz = np.random.randint(0, 100, size=(5, 5))
print("Matriz:")
print(matriz)


media = np.mean(matriz)
desviacion_estandar = np.std(matriz)

print("Media:", media)
print("Desviación estándar:", desviacion_estandar)


plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Valores')
plt.title('Matriz 5x5')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.xticks(ticks=np.arange(5), labels=np.arange(5))
plt.yticks(ticks=np.arange(5), labels=np.arange(5))


plt.show()  
```

```python
import numpy as np
import matplotlib.pyplot as plt

matriz = np.random.randint(0, 10, size=(5, 5))
print("Matriz:")
print(matriz)

media = np.mean(matriz)
desviacion_estandar = np.std(matriz)

print("Media:", media)
print("Desviación estándar:", desviacion_estandar)

plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Valores')
plt.title('Matriz 5x5')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.xticks(ticks=np.arange(5), labels=np.arange(5))
plt.yticks(ticks=np.arange(5), labels=np.arange(5))

# Anotar la media y desviación estándar en la gráfica

plt.show()

```
